### Simple Moving Averages (SMA) Calculation (Lagging Indicator) <a id='1.4.1_Simple_Moving_Averages'></a>

As per [Investopedia](https://www.investopedia.com/ask/answers/042815/how-simple-moving-average-calculated.asp), the **SMA indicator is used to indicate buy and sell signals** to traders and investors. It helps to identify support and resistance prices of stocks to signal where the asset should be traded. Traders and investors also use SMA crossovers to indicate bullish and bearish price action.

The formula for the SMA is as follow:


```python
# Calculating weekly SMA for the SPY dataset
spy['sma5'] = ta.SMA(np.asarray(spy['close']), 5) # Since trading is done for 5 days only, the weekly value is 5.

# Plotting the weekly SMA for the SPY
spy.plot(y=['close', 'sma5'], figsize=(20,6))
plt.title("SPY close prices and Weekly SMA")
plt.legend(loc='best')
plt.show()
```

### Exponential Moving Averages (EMA) Calculation (Lagging Indicator) <a id='1.4.2_Exponential_Moving_Averages'></a>

As per [Investopedia](https://www.investopedia.com/ask/answers/042815/how-simple-moving-average-calculated.asp), an exponential moving average (EMA) is a type of moving average (MA) that places a greater weight and significance on the most recent data points. The exponential moving average is also referred to as the exponentially weighted moving average. An exponentially weighted moving average reacts more significantly to recent price changes than a simple moving average (SMA), which applies an equal weight to all observations in the period. 

Key features of EMA are:

- The EMA is a moving average that places a greater weight and significance on the most recent data points.
- Like all moving averages, this technical indicator is used to produce buy and sell signals based on crossovers and divergences from the historical average.
- Traders often use several different EMA lengths, such as 10-day, 50-day, and 200-day moving averages.

 The Formula for EMA Is:



```python
# Calculating EMA for the SPY dataset for 10-day, 50-day and 200-day moving averages
spy['ema10'] = ta.EMA(np.asarray(spy['close']), 10)
spy['ema50'] = ta.EMA(np.asarray(spy['close']), 50)
spy['ema200'] = ta.EMA(np.asarray(spy['close']), 200)

# Plotting for 10-day, 50-day and 200-day EMA for the SPY
spy.plot(y=['close', 'ema10', 'ema50', 'ema200'],figsize=(20,6))
plt.title("SPY Close Price & EMA Plot for 10-day, 50-day and 200-day")
plt.legend(loc='upper left')
plt.show()
```

### Bollinger Bands Indicator Calculation (Lagging Indicator) <a id='1.4.3_Bollinger_Bands_Indicator'></a>

As per [Investopedia](https://www.investopedia.com/terms/b/bollingerbands.asp), Bollinger BandsÂ® were developed and copyrighted by famous technical trader John Bollinger, designed to discover opportunities that give investors a higher probability of properly identifying when an asset is oversold or overbought.

The formula for the Bollinder Band is as follow:


Bollinger BandsÂ® consist of a centerline and two price channels (bands) above and below it. The centerline is an exponential moving average; the price channels are the standard deviations of the stock being studied. The bands will expand and contract as the price action of an issue becomes volatile (expansion) or becomes bound into a tight trading pattern (contraction). 

```python
# Calculating Bollinger bands for the SPY dataset
spy['upper'], spy['middle'], spy['lower'] = ta.BBANDS(np.asarray(spy['close']), timeperiod=20, 
                                                                 nbdevup=2, nbdevdn=2, matype=0)
# Plotting the charts for the Bollinger bands
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(spy['close'], label='Close')
ax.plot(spy['upper'], linestyle='--', label='Upper')
ax.plot(spy['middle'], linestyle='--', label='Middle')
ax.plot(spy['lower'], linestyle='--', label='Lower')
plt.xlabel("Timeperiod")
plt.ylabel("Closing Price of SPY")
plt.title("SPY close price and Bollinger Bands")
plt.legend(loc='upper left')
plt.show()
```

### Parabolic Stop and Reverse (SAR) Indicator Calculation (Lagging Indicator) <a id='1.4.4_Parabolic_SAR_Indicator'></a>

As per [Investopedia](https://www.investopedia.com/terms/p/parabolicindicator.asp), the parabolic SAR indicator, developed by J. Wells Wilder, is used by traders to determine trend direction and potential reversals in price. The indicator uses a trailing stop and reverse method called "SAR," or stop and reverse, to identify suitable exit and entry points. 

The parabolic indicator generates buy or sell signals when the position of the dots moves from one side of the asset's price to the other. For example, a buy signal occurs when the dots move from above the price to below the price, while a sell signal occurs when the dots move from below the price to above the price.

```python
# Calculating Parabolic SAR for the SPY dataset
spy['sar'] = ta.SAR(np.asarray(spy['high']), np.asarray(spy['low']), acceleration=0.02, maximum=0.2)

# Plotting SAR for the SPY dataset
fig, ax = plt.subplots(figsize=(20,10))
ax.plot(spy['close'], label='Close')
ax.plot(spy['sar'], marker='o', linestyle='', label='SAR')
plt.title("SPY close price and Parabolic SAR")
plt.legend(loc='upper left')
plt.show()
```

### Average Directional Movement Index (ADX) Calculation (Leading Indicator) <a id='1.5.1_Avg_Directional_Movement_Index'></a>

As per [Investopedia](https://www.investopedia.com/ask/answers/112814/how-average-directional-index-adx-calculated-and-what-formula.asp), the Average Directional Index (ADX) is used to indicate market direction, the existence or nonexistence of a trend and market momentum. 

Market direction is determined by the levels of the +DI (positive Directional Indicator) and -DI (negative Directional Indicator). If +DI is the higher number, market direction is up; if -DI is the greater number, market direction is down. The ADX indicator, which varies in value from zero to 100, is the primary momentum indicator. A value over 20 indicates the existence of a trend; a value over 40 indicates a strong trend. 

```python
# Calculating ADX for the SPY dataset for 14 days smoothing
spy['adx'] = ta.ADX(np.asarray(spy['high']), np.asarray(spy['low']), np.asarray(spy['close']), timeperiod=14)
spy['+di'] = ta.PLUS_DI(np.asarray(spy['high']), np.asarray(spy['low']), np.asarray(spy['close']), timeperiod=14)
spy['-di'] = ta.MINUS_DI(np.asarray(spy['high']), np.asarray(spy['low']), np.asarray(spy['close']), timeperiod=14)

# Plotting ADX Indicator chart
fig, ax = plt.subplots(2, sharex=True, figsize=(20,12))
ax[0].plot(spy['close'], label='Close')
ax[0].legend(loc='upper left')
ax[1].plot(spy['adx'], color='k', label='ADX')
ax[1].plot(spy['+di'], color='g', label='+ DI')
ax[1].plot(spy['-di'], color='r', label='- DI')
ax[1].legend(loc='upper left')
plt.suptitle("Close price for SPY and ADX Indicator for 14 days")
plt.show()
```

### Commodity Channel Index (CCI) Calculation (Leading Indicator) <a id='1.5.2_Commodity_Channel_Index'></a>

As per [Investopedia](https://www.investopedia.com/terms/c/commoditychannelindex.asp), the Commodity Channel Indexâ€‹ (CCI) is a momentum-based oscillator used to help determine when an investment vehicle is reaching a condition of being overbought or oversold. When the CCI is above zero, it indicates the price is above the historic average. Conversely, when the CCI is below zero, the price is below the historic average.



```python
# Calculating CCI for the SPY dataset with smoothing for 20 days period (P)
spy['cci'] = ta.CCI(np.asarray(spy['high']), np.asarray(spy['low']), np.asarray(spy['close']), timeperiod=20)

# Plotting CCI chart for the SPY dataset
fig, ax = plt.subplots(2, sharex=True, figsize=(20,12))
ax[0].plot(spy['close'], label='Close')
ax[0].legend(loc='upper left')
ax[1].plot(spy['cci'], label='CCI', color='green')
ax[1].legend(loc='upper left')
plt.suptitle("SPY Close Price and CCI Chart for 20-days period")
plt.show()
```

### Momentum Average Convergence/Divergence (MACD) Calculation (Leading Indicator) <a id='1.5.3_MACD_Calculation'></a>

As per [Investopedia](https://www.investopedia.com/terms/m/macd.asp), moving average convergence divergence (MACD) is a trend-following momentum indicator that shows the relationship between two moving averages of a securityâ€™s price. Moving average convergence divergence (MACD) is calculated by subtracting the 26-period exponential moving average (EMA) from the 12-period EMA.

MACD triggers technical signals when it crosses above (to buy) or below (to sell) its signal line. The speed of crossovers is also taken as a signal of a market is overbought or oversold. 

MACD helps investors understand whether the bullish or bearish movement in the price is strengthening or weakening.

The formula is as follow:


```python
# Calculating MACD for 12-day short-term smoothing, 26-day long-term smoothing and signal period of 9 days
spy['macd'], spy['macdema'], spy['macdhist'] = ta.MACD(np.asarray(spy['close']), 
                                                       fastperiod=12, slowperiod=26, signalperiod=9)

# Plotting MACD chart for the SPY dataset
fig, ax = plt.subplots(2, sharex=True, figsize=(20,12))
ax[0].plot(spy['close'], label='Close')
ax[0].legend(loc='upper left')
ax[1].plot(spy['macd'], label='MACD')
ax[1].plot(spy['macdema'], label='MACD EMA', color='green')
ax[1].plot(spy['macdhist'], label='MACDHIST', linestyle='--')
ax[1].legend(loc='upper left')
plt.suptitle("SPY Close Price and MACD Chart")
plt.show()
```

### Rate of Change (ROC) Indicator Calculation (Leading Indicator) <a id='1.5.4_Rate_Of_Change'></a>

As per [Investopedia](https://www.investopedia.com/terms/r/rateofchange.asp), rate of change (ROC) is used to mathematically describe the percentage change in value over a defined period of time, and it represents the momentum of a variable. The calculation for ROC is simple in that it takes the current value of a stock or index and divides it by the value from an earlier period. Subtract one and multiply the resulting number by 100 to give it a percentage representation. 

Here is the formula for the ROC calculation:


```python
# Calculating ROC for the SPY dataset with 21-day
spy['roc'] = ta.ROC(np.asarray(spy['close']), timeperiod=21)

# Plotting ROC chart for the SPY dataset
fig, ax = plt.subplots(2, sharex=True, figsize=(20,10))
ax[0].plot(spy['close'], label='Close')
ax[0].legend(loc='upper left')
ax[1].plot(spy['roc'], label='ROC', color='green')
ax[1].legend(loc='upper left')
plt.suptitle("SPY Close Price and the ROC Chart")
plt.show()
```

### Relative Strength Indicator (RSI) Calculation (Leading Indicator) <a id='1.5.5_RSI_Calculation'></a>

As per [Investopedia](https://www.investopedia.com/terms/r/rsi.asp), the relative strength index (RSI) is a momentum indicator used in technical analysis that measures the magnitude of recent price changes to evaluate overbought or oversold conditions in the price of a stock or other asset. 

The RSI is displayed as an oscillator (a line graph that moves between two extremes) and can have a reading from 0 to 100. An asset is usually considered overbought when the RSI is above 70% and oversold when it is below 30%.



```python
# Calculating RSI for SPY dataset for 14-day period
spy['rsi'] = ta.RSI(np.asarray(spy['close']), timeperiod=14)

# Plotting RSI chart for the SPY dataset
fig, ax = plt.subplots(2, sharex=True, figsize=(20,10))
ax[0].plot(spy['close'], label='Close')
ax[0].legend(loc='upper left')
ax[1].plot(spy['rsi'], label='RSI', color='green')
ax[1].legend(loc='upper left')
plt.suptitle("SPY Close Price and RSI Chart for 14-day Period")
plt.show()
```

### Stochastic Oscillator Calculation (Leading Indicator) <a id='1.5.6_Stochastic_Oscillator'></a>

As per [Investopedia](https://www.investopedia.com/terms/s/stochasticoscillator.asp), a stochastic oscillator is a momentum indicator comparing a particular closing price of a security to a range of its prices over a certain period of time. The sensitivity of the oscillator to market movements is reducible by adjusting that time period or by taking a moving average of the result. It is used to generate overbought and oversold trading signals, utilizing a 0â€“100 bounded range of values. 



```python
# Calculating Stochastic Oscillator for SPY dataset (finding 'slowk' and 'slowd')
spy['slowk'], spy['slowd'] = ta.STOCH(np.asarray(spy['high']), np.asarray(spy['low']), np.asarray(spy['close']),
                                     fastk_period=14, slowk_period=3, slowk_matype=0, slowd_period=3, 
                                      slowd_matype=0)

# Plotting the Stochastic Oscillator chart for the SPY dataset 
fig, ax = plt.subplots(2, sharex=True, figsize=(20,12))
ax[0].plot(spy['close'], label='Close')
ax[0].legend(loc='upper left')
ax[1].plot(spy['slowk'], color='red', linestyle='--', label='SLOWK')
ax[1].plot(spy['slowd'], color='green', label='SLOWD')
ax[1].legend(loc='upper left')
plt.suptitle("SPY Close Price and Stochastic Oscillator Chart")
plt.show()
```

### Williams %R Indicator Calculation (Leading Indicator) <a id='1.5.7_WilliamsR_Indicator'></a>

### Trade with SMA and EMA Indicators for 'TSLA' Ticker  <a id='1.6.2_SMA_EMA_Trading'></a>

As per [Investopedia](https://www.investopedia.com/articles/active-trading/052014/how-use-moving-average-buy-stocks.asp), the most common applications of moving averages are to identify trend direction and to determine support and resistance levels. 

As a general guideline, if the price is above a moving average (MA), the trend is up. If the price is below a moving average, the trend is down. However, moving averages can have different lengths, so one MA may indicate an uptrend while another MA indicates a downtrend. 

Common moving average lengths are 10, 20, 50, 100 and 200. The 20-day may be of analytical benefit to a shorter-term trader since it follows the price more closely and therefore produces less "lag" than the longer-term moving average. A 100-day MA may be more beneficial to a longer-term trader. 

Another popular type of moving average is the exponential moving average (EMA). The calculation is more complex, as it applies more weighting to the most recent prices. If you plot a 50-day SMA and a 50-day EMA on the same chart, you'll notice that the EMA reacts more quickly to price changes than the SMA does, due to the additional weighting on recent price data. 

**Trading Strategiesâ€”Crossovers**

Crossovers are one of the main moving average strategies. The first type is a price crossover, which is when the price crosses above or below a moving average to signal a potential change in trend. 

Another strategy is to apply two moving averages to a chart: one longer and one shorter. When the shorter-term MA crosses above the longer-term MA, it's a buy signal, as it indicates that the trend is shifting up. This is known as a "golden cross."

Meanwhile, when the shorter-term MA crosses below the longer-term MA, it's a sell signal, as it indicates that the trend is shifting down. This is known as a "dead/death cross." 

```python
# Calculating SMA(5) for 5-day and chart it
tsla['sma5'] = ta.SMA(np.asarray(tsla['Close']), 5)

# Calculating price "Single Crossover Trading" SMA with one day back prices and 2-day back prices

tsla['Close(-1)'] = tsla['Close'].shift(1)
tsla['sma5(-1)'] = tsla['sma5'].shift(1)
tsla['Close(-2)'] = tsla['Close'].shift(2)
tsla['sma5(-2)'] = tsla['sma5'].shift(2)

# Identifying trading signals as buy=1, sell=-1, do nothing=0
tsla['sma5signal'] = 0
sma5signal = 0
for i, j in enumerate(tsla.iterrows()):
    if j[1]['Close(-2)'] < j[1]['sma5(-2)'] and j[1]['Close(-1)'] > j[1]['sma5(-1)']:
        sma5signal=1
    elif j[1]['Close(-2)'] > j[1]['sma5(-2)'] and j[1]['Close(-1)'] < j[1]['sma5(-1)']:
        sma5signal= -1
    else:
        sma5signal=0
    tsla.iloc[i, 11] = sma5signal # last column number 11 to store the sma5signal
    
# Plotting trading signals chart
fig, ax = plt.subplots(2, sharex=True, figsize=(20,12))
ax[0].plot(tsla['Close'], label='Close')
ax[0].plot(tsla['sma5'], label='SMA5', color='red')
ax[0].legend(loc='upper left')
ax[1].plot(tsla['sma5signal'], marker='o', linestyle='', label='SMA5SIGNAL')
ax[1].legend(loc='best')
plt.suptitle("TSLA Close Price and SMA5 Chart")
plt.show()
```

If the value of the 'SMA5SIGNAL' is 1, it means it is time to buy the TSLA stock and if it is -1, it is time to sell the stock.

```python
# Calculating EMA (Exponential Moving Averages) price for 5-day and 21-day averages
tsla['ema5'] = ta.EMA(np.asarray(tsla['Close']), 5)
tsla['ema21'] = ta.EMA(np.asarray(tsla['Close']), 21)

# Calculating price "Double Crossover Trading" EMA with 1-day back prices & 2-day back prices
tsla['ema5(-1)'] = tsla['ema5'].shift(1)
tsla['ema21(-1)'] = tsla['ema21'].shift(1)
tsla['ema5(-2)'] = tsla['ema5'].shift(2)
tsla['ema21(-2)'] = tsla['ema21'].shift(2)

# Identifying trading signals as buy=1, sell=-1, do nothing=0
tsla['emasignal'] = 0
emasignal = 0
for i, j in enumerate(tsla.iterrows()):
    if j[1]['ema5(-2)'] < j[1]['ema21(-2)'] and j[1]['ema5(-1)'] > j[1]['ema21(-1)']:
        emasignal=1
    elif j[1]['ema5(-2)'] > j[1]['ema21(-2)'] and j[1]['ema5(-1)'] < j[1]['ema21(-1)']:
        emasignal= -1
    else:
        emasignal=0
    tsla.iloc[i, 18] = emasignal # last column number 18 to store the emasignal
    
# Plotting trading signals chart
fig, ax = plt.subplots(2, sharex=True, figsize=(20,12))
ax[0].plot(tsla['Close'], label='Close')
ax[0].plot(tsla['ema5'], label='EMA5', color='red')
ax[0].plot(tsla['ema21'], label='EMA21', color='green')
ax[0].legend(loc='upper left')
ax[1].plot(tsla['emasignal'], marker='o', linestyle='', label='EMASIGNAL')
ax[1].legend(loc='best')
plt.suptitle("TSLA Close Price with EMA5, EMA21 and EMA Signal Chart")
plt.show()
```

If the value of the 'EMASIGNAL' is 1, it means it is time to buy the TSLA stock and if it is -1, it is time to sell the stock.

### Trade with BB and SAR Indicators for 'TSLA' Ticker  <a id='1.6.3_BB_SAR_Trading'></a>

Bollinger Bands are used to identify if market is overbought or oversold. Parabolic SAR provides reversible signal. When market price reverses from overbought and oversold area, parabolic SAR confirms the new trend. 

**Buy signal**: When market reaches to lower Bollinger Bands, it is considered in the oversold area. When the dots of Parabolic SAR appear below the price, then it confirms buy signal. 

**Sell signal**: Upper Bollinger Bands indicates overbought area. When market rejects from upper bands and the dots of Parabolic SAR appear above the market price, then it's a sell signal.

```python
# Calculating Bollinger Bands
tsla['upper'], tsla['middle'], tsla['lower'] = ta.BBANDS(np.asarray(tsla['Close']), 
                                                         timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
# Calculating BB Crossover chart
tsla['Close(-1)'] = tsla['Close'].shift(1)
tsla['upper(-1)'] = tsla['upper'].shift(1)
tsla['lower(-1)'] = tsla['lower'].shift(1)
tsla['Close(-2)'] = tsla['Close'].shift(2)
tsla['upper(-2)'] = tsla['upper'].shift(2)
tsla['lower(-2)'] = tsla['lower'].shift(2)

# Identifying trading signals as buy=1, sell=-1, do nothing=0
tsla['bbsignal'] = 0
bbsignal = 0
for i, j in enumerate(tsla.iterrows()):
    if j[1]['Close(-2)'] < j[1]['lower(-2)'] and j[1]['Close(-1)'] > j[1]['lower(-1)']:
        bbsignal=1
    elif j[1]['Close(-2)'] < j[1]['upper(-2)'] and j[1]['Close(-1)'] > j[1]['upper(-1)']:
        bbsignal= -1
    else:
        bbsignal=0
    tsla.iloc[i, 26] = bbsignal # last column number 26 to store the bbsignal
    
# Plotting BB Chart
fig, ax = plt.subplots(2, sharex=True, figsize=(20,12))
ax[0].plot(tsla['Close'], label='Close')
ax[0].plot(tsla['upper'], linestyle='--', label='Upper Band')
ax[0].plot(tsla['middle'], linestyle='--', label='Middle Band')
ax[0].plot(tsla['lower'], linestyle='--', label='Lower Band')
ax[0].legend(loc='upper left')
ax[1].plot(tsla['bbsignal'], marker='o', linestyle='', label='BBSIGNAL')
ax[1].legend(loc='upper left')
plt.suptitle("TSLA Close Price and Bollinger Bands")
plt.show()
```

If the value of the 'BBSIGNAL' is 1, it means it is time to buy the TSLA stock and if it is -1, it is time to sell the stock.

```python
# Calculating parabolic SAR for the TSLA ticker
tsla['sar'] = ta.SAR(np.asarray(tsla['High']), np.asarray(tsla['Low']), 
                     acceleration=0.02, maximum=0.2)

# Calculating SAR Crossover chart
tsla['Close(-1)'] = tsla['Close'].shift(1)
tsla['sar(-1)'] = tsla['sar'].shift(1)
tsla['Close(-2)'] = tsla['Close'].shift(2)
tsla['sar(-2)'] = tsla['sar'].shift(2)

# For the trading signal buy=1, sell=-1, do nothing=0
tsla['sarsignal'] = 0
sarsignal = 0
for i, j in enumerate(tsla.iterrows()):
    if j[1]['Close(-2)'] < j[1]['sar(-2)'] and j[1]['Close(-1)'] > j[1]['sar(-1)']:
        sarsignal=1
    elif j[1]['Close(-2)'] > j[1]['sar(-2)'] and j[1]['Close(-1)'] < j[1]['sar(-1)']:
        sarsignal= -1
    else:
        sarsignal=0
    tsla.iloc[i, 30] = sarsignal # last column number 30 to store the sarsignal
    
# Plotting SAR Chart
fig, ax = plt.subplots(2, sharex=True, figsize=(20,12))
ax[0].plot(tsla['Close'], label='Close')
ax[0].plot(tsla['sar'], linestyle='', marker='o', label='SAR')
ax[0].legend(loc='upper left')
ax[1].plot(tsla['sarsignal'], marker='o', linestyle='', label='SARSIGNAL')
ax[1].legend(loc='upper left')
plt.suptitle("TSLA Close Price and SAR Signals")
plt.show()
```

If the value of the 'SARSIGNAL' is 1, it means it is time to buy the TSLA stock and if it is -1, it is time to sell the stock.

### Trade with MACD and RSI Indicators for 'TSLA' Ticker  <a id='1.6.4_MACD_RSI_Trading'></a>

As per [Investopedia](https://www.investopedia.com/ask/answers/122214/what-are-main-differences-between-moving-average-convergence-divergence-macd-relative-strength-index.asp), the MACD measures the relationship between two EMAs, while the RSI measures price change in relation to recent price highs and lows. These two indicators are often used together to provide analysts a more complete technical picture of a market. 

The RSI aims to indicate whether a market is considered to be overbought or oversold in relation to recent price levels. The RSI calculates average price gains and losses over a given period of time; the default time period is 14 periods with values bounded from 0 to 100. 

```python
# Calculating MACD for 12-day short-term smoothing, 26-day long-term smoothing and signal period of 9 days
tsla['macd'], tsla['macdema'], tsla['macdhist'] = ta.MACD(np.asarray(tsla['Close']), 
                                                       fastperiod=12, slowperiod=26, signalperiod=9)

# Calculating SAR Crossover chart
tsla['macd(-1)'] = tsla['macd'].shift(1)
tsla['macdema(-1)'] = tsla['macdema'].shift(1)
tsla['macd(-2)'] = tsla['macd'].shift(2)
tsla['macdema(-2)'] = tsla['macdema'].shift(2)

# For the trading signal buy=1, sell=-1, do nothing=0
tsla['macdsignal'] = 0
macdsignal = 0
for i, j in enumerate(tsla.iterrows()):
    if j[1]['macd(-2)'] < j[1]['macdema(-2)'] and j[1]['macd(-1)'] > j[1]['macdema(-1)']:
        macdsignal=1
    elif j[1]['macd(-2)'] > j[1]['macdema(-2)'] and j[1]['macd(-1)'] < j[1]['macdema(-1)']:
        macdsignal= -1
    else:
        macdsignal=0
    tsla.iloc[i, 38] = macdsignal # last column number 38 to store the macdsignal
    
# Plotting MACD Chart
fig, ax = plt.subplots(3, sharex=True, figsize=(20,18))
ax[0].plot(tsla['Close'], label='Close')
ax[0].legend(loc='upper left')
ax[1].plot(tsla['macd'], label='MACD')
ax[1].plot(tsla['macdema'], label='MACDEMA')
ax[1].plot(tsla['macdhist'], linestyle='--', label='MACDHIST')
ax[1].legend(loc='upper left')
ax[2].plot(tsla['macdsignal'], marker='o', linestyle='', label='MACDSIGNAL')
ax[1].legend(loc='upper left')
plt.suptitle("TSLA Close Price and MACD Signals")
plt.show()
```

If the value of the 'MACDSIGNAL' is 1, it means it is time to buy the TSLA stock and if it is -1, it is time to sell the stock.

```python
# Calculating RSI for 14-day period
tsla['rsi'] = ta.RSI(np.asarray(tsla['Close']), timeperiod=14)

# Calculating SAR Crossover chart
tsla['rsi(-1)'] = tsla['rsi'].shift(1)
tsla['rsi(-2)'] = tsla['rsi'].shift(2)

# For the trading signal buy=1, sell=-1, do nothing=0
tsla['rsisignal'] = 0
rsisignal = 0
for i, j in enumerate(tsla.iterrows()):
    if j[1]['rsi(-2)'] < 30 and j[1]['rsi(-1)'] > 30:
        rsisignal=1
    elif j[1]['rsi(-2)'] < 70 and j[1]['rsi(-1)'] > 70:
        rsisignal= -1
    else:
        rsisignal=0
    tsla.iloc[i, 42] = rsisignal # last column number 42 to store the rsisignal
    
# Plotting RSI Chart
fig, ax = plt.subplots(3, sharex=True, figsize=(20,18))
ax[0].plot(tsla['Close'], label='Close')
ax[0].legend(loc='upper left')
ax[1].plot(tsla['rsi'], label='RSI', color='green')
ax[1].axhline(y=30, linestyle='--', color='red')
ax[1].axhline(y=70, linestyle='--', color='red')
ax[1].legend(loc='upper left')
ax[2].plot(tsla['rsisignal'], marker='o', linestyle='', color='orange', label='RSISIGNAL')
ax[1].legend(loc='upper left')
plt.suptitle("TSLA Close Price and RSI Signals")
plt.show()
```

If the value of the 'RSISIGNAL' is 1, it means it is time to buy the TSLA stock and if it is -1, it is time to sell the stock.
