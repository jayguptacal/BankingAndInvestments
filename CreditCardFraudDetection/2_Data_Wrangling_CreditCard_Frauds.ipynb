# Data Wrangling

The focus of this Jupyter notebook is on the data wrangling steps which are as follows:

- Identification of **rows and columns** and **data types**     
- Checking and imputation of **missing values** if required in this notebook or during the EDA phase  
- Checking of **usefulness of required column features** using corr() and heatmap() functions to decide which column features to keep and which ones to drop
- Based on the above analysis, keep the required features of interest and save it for the next phase
     
In summary, the focus of this step is on exploring and analyzing the data to decide on the steps required for the feature engineering in the next notebook.

## Importing Packages and Libraries<a id='2.1_Imports'></a>

```python
# Imports of required packages and libraries for the data wrangling purpose only
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
import warnings
warnings.simplefilter(action='ignore')
```

## Loading the Dataset<a id='2.2_Loading_Data'></a>

```python
# Loading the dataset from the desktop
data = pd.read_csv('creditcard.csv')

data.head()
```

```python
# Printing the number of rows and columns
print("Total number of rows are {} and columns are {}.".format(data.shape[0], data.shape[1]))
```

```python
# Printing data types
data.info()
```

## Checking for Missing Values<a id='2.3_Checking_Missing_Values'></a>

```python
# Checking for missing values
data.isnull().sum()
```

There are no missing values in the dataset.

```python
# Calculating describe()
data.describe().T
```

On visual checking, it is clear that there are many columns here but not sure yet if all the columns are needed. For sure I would need 'Time', 'Amount' and 'Class' for further analysis.

## Checking for Relationships<a id='2.4_Checking_Relationships'></a>

```python
# Checking for the relationships
data.corr()
```

```python
# Plotting the heatmap for better visual representation
plt.figure(figsize=(15,15))
sns.heatmap(data.corr(), annot=True, fmt='.2f', cmap='viridis')
plt.tight_layout()
plt.show()
```

From the above heatmap, it seems that I can drop all the columns except for the 'Time', 'Amount' and 'Class'.

```python
# Filtering for the final required features of interest
df = data.loc[:, ['Time', 'Amount', 'Class']]
df.head()
```

```python
# Checking again for the data
df.dtypes
```

```python
# Checking with describe()
df.describe().T
```

```python
# Printing for the rows and columns for the filtered data
print("Total number of rows are {} and columns are {}.".format(df.shape[0], df.shape[1]))
```

```python
# Saving this new dataset for the EDA phase
df.to_csv('df_eda.csv', index=False)
```
