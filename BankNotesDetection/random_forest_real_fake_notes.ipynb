# Project Title: Predict Genuine vs. Fake Bank Notes based on Random Forest Classifier Modeling, Supervised Machine Learning Algorithm

## Project Background

This dataset was developed using genuine and fake bank notes which were scanned and wavelet transformation was performed on the images. During this process, following features were collected:

- Variance of wavelet transformation
- Skewness of wavelet transformation
- Curtosis of wavelet transformation
- Entropy of image

If you are wondering, what is wavelet transformation?

As per Science Direct (https://www.sciencedirect.com/topics/computer-science/wavelet-transforms), wavelets are a general way to analyze multi-dimensional images. This is very useful for image compression and removing noise. If you are looking for more information, you can check out at the Wikipedia at https://en.wikipedia.org/wiki/Discrete_wavelet_transform.

**The goal of this project is to use Random Forest Classifier, a supervised machine learning algorithm to predict genuine vs fake bank notes.**

## Project Methodology

# Problem Identification

**Hypothesis:** Identify and predict genuine vs fake bank notes utilizing Random Forest Classifier, a supervised machine learning algorithm based on various bank note's features.

**Context for bank notes Identification:** As part of the scientific study, the goal is to develop a model which can identify fake bank notes with higher degree of confidence thereby minimizing losses for the banks.

# Pre-Processing and Training Data Development

```python
# Importing required packages
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Class', axis=1)
```

```python
y = df['Class']
```

```python
# Splitting data into training and testing size of 20%
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=101)
```

# Machine Learning Modeling

## Random Forest Classifier with GridSearchCV

```python
# Importing required packages
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score
```

```python
# Identifying random forest classifier hyperparameters
# help(RandomForestClassifier)
```

**Hyperparameters:**

RandomForestClassifier(n_estimators=100, *, criterion='gini', max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None, min_impurity_decrease=0.0, min_impurity_split=None, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False, class_weight=None, ccp_alpha=0.0, max_samples=None)

```python
# Mainly playing with n_estimators, max_features, bootstrap, and oob_score as these diretly impact random forest
n_estimators = [50,100,150,200,250,300]
max_features = [2,3,4] # there are only 4 features available here
bootstrap = [True,False]
oob_score = [True,False] # out of bag will be possible only when bootstrap is true, otherwise it will give warning.
```

```python
# setting up the parametes for the grid
param_grid = {"n_estimators": n_estimators,
             "max_features": max_features,
             "bootstrap": bootstrap,
             "oob_score": oob_score}
```

```python
# Creating grid now
rfc = RandomForestClassifier()
grid = GridSearchCV(rfc,param_grid)
```

```python
# Fitting training data to grid
grid.fit(X_train,y_train)
```

```python
# Finding best parameters
grid.best_params_
```

```python
# creating a model with these best parameters
rfc_best = RandomForestClassifier(n_estimators=50, max_features=2,oob_score=True)
```

```python
# fitting training data to this new model
rfc_best.fit(X_train,y_train)
```

```python
# Checking for oob_score
rfc_best.oob_score_
```

```python
# Predicting with this model
preds_best = rfc_best.predict(X_test)
```

```python
# printing classification report
print("Classification Report for Optimized Random Forest Classifier:\n",classification_report(y_test,preds_best))
```

```python
# plotting confusion matrix
plot_confusion_matrix(rfc_best,X_test,y_test);
```

**Based on the above, only 4 are mis-classified.**

## Random Forest Classifier for different Number of Estimators

```python
# From the above, we can fix max_features=2 and now we can start playing with number of estimators to see the best fit

errors = []
misclassifications = []

for n in range(1,200):
    rfc = RandomForestClassifier(n_estimators=n,max_features=2)
    rfc.fit(X_train,y_train)
    
    preds = rfc.predict(X_test)
    err = 1 - accuracy_score(y_test,preds)
    miss_classifieds = np.sum(preds != y_test)
    errors.append(err)
    misclassifications.append(miss_classifieds)
    
```

```python
# plotting for number of estimators and error
plt.figure(figsize=(8,6))
plt.plot(range(1,200),errors)
plt.title("Plot between Number of Estimators vs. Error")
plt.xlabel("n_estimators")
plt.ylabel("Error")
plt.show();
```

```python
# plotting for number of estimators and missclassifications
plt.figure(figsize=(8,6))
plt.plot(range(1,200),misclassifications)
plt.title("Plot between Number of Estimators vs. misclassifications")
plt.xlabel("n_estimators")
plt.ylabel("misclassifications")
plt.show();
```

From this test, it seems that we don't need even 50 estimators and we can get the same results with n_estimators=25.

## Random Forest Classifier with n_estimators=25

```python
# running the model for n_estimators=25
rfc_25 = RandomForestClassifier(n_estimators=25, max_features=2)
rfc_25.fit(X_train,y_train)
preds_25 = rfc_25.predict(X_test)
```

```python
# Classification report for n_estimators=25
print("RFC Classification Report with n_estimators=25:\n",classification_report(y_test,preds_25))
```

```python
# plotting confusion matrix for n_estimators = 25
plot_confusion_matrix(rfc_25,X_test,y_test);
```

Computing is expensive, so using n_estimators=25 now. Also trying for different max_features=3.

## Random Forest Classifier with max_features=3

```python
# running another model with max_features=3
rfc_3 = RandomForestClassifier(n_estimators=25,max_features=3)
rfc_3.fit(X_train,y_train)
preds_3 = rfc_3.predict(X_test)
```

```python
# Printing classification report for max_features=3
print("RFC Classification Report with max_features=3:\n",classification_report(y_test,preds_3))
```

```python
# plotting confusion matrix for max_features = 3
plot_confusion_matrix(rfc_3,X_test,y_test);
```

**There is some improvement here as only 4 are mis-classified now.**
